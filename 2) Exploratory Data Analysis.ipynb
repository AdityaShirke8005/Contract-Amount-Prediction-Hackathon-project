```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Final_contract_data.csv')
df = df.drop('Unnamed: 0',axis=1)
df.head()
```

```python
df.describe()
```

### Feature distributions using Seaborn

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='year', color='orange')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Distribution of Years')
plt.show()
```

We have a lot of data for the year 2021 and 2010

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['GDP_growth'], kde=True, color='pink', edgecolor='black')
plt.xlabel('GDP Growth')
plt.ylabel('Frequency')
plt.title('Histogram of GDP Growth')
plt.show()
```

GDP growth have approximately normal distributions, which is expected for macroeconomic indicators.

```python
# Histogram for 'inflation'
plt.figure(figsize=(8, 6))
sns.histplot(df['inflation'], kde=True, color='purple', edgecolor='black')
plt.xlabel('Inflation')
plt.ylabel('Frequency')
plt.title('Histogram of Inflation')
plt.show()
```

Inflation rate is quite frequent and stable in the range of [1 ,3] 

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['manufacturing_value_added'], kde=True,color='gold', edgecolor='black')
plt.xlabel('Manufacturing Value Added')
plt.ylabel('Frequency')
plt.title('Histogram of Manufacturing Value Added')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['official_exchange_rate'], kde=True, color='red', edgecolor='black')
plt.xlabel('Official Exchange Rate')
plt.ylabel('Frequency')
plt.title('Histogram of Official Exchange Rate')
plt.show()
```

official exchange rate has a relatively constant distribution, indicating a stable exchange rate over the years.

```python
sns.pairplot(df[['original_amount', 'revised_amount', 'GDP_growth', 'inflation', 'manufacturing_value_added']])
plt.show()
```

```python
correlation_matrix = df[['original_amount', 'revised_amount', 'GDP_growth', 'inflation', 'manufacturing_value_added', 'official_exchange_rate']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

## Valuable Insights

- There is a **positive correlation** between 'original_amount' and 'revised_amount', which is expected as the **revised amount is likely to be higher than the original amount in most cases.**

- GDP_growth and inflation show a moderate negative correlation, which is in line with standard economic theory **where high inflation can negatively impact GDP growth.**

- Other numeric features show weak correlations, indicating limited linear relationships between them.
